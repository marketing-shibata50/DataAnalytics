```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
db = pd.read_excel('data.xlsx', sheet_name='Liny', header=1)
db = db[['登録(フォロー)日時', 'ユーザーブロック', '流入時シナリオ', '希望勤務地', '個人情報入力経路', '流入経路', '流入経路詳細', '卒業年度', '電話番号']]
```

```python
db['登録(フォロー)日時'] = db['登録(フォロー)日時'].dt.strftime('%Y%m%d')
db['登録(フォロー)日時'] = db['登録(フォロー)日時'].astype('datetime64')
```

```python
db['登録(フォロー)日時'].min()
```

```python
s_time = dt.datetime(2021,3,1)
e_time = dt.datetime(2021,3,31)
temp_db = db[(db['登録(フォロー)日時'] >= s_time) & (db['登録(フォロー)日時'] <= e_time)]
```

```python
tag = db['流入経路詳細'].unique()
```

```python
data = pd.pivot_table(data = temp_db, values = ["ユーザーブロック", "卒業年度", "電話番号"], aggfunc ="count", index = "登録(フォロー)日時", margins=True, fill_value=0)
data = data.rename(columns={'ユーザーブロック': '友だち', '電話番号': '個人情報'})
```

```python
data['卒業年度%'] = (data['卒業年度'] / data['友だち']).map('{:.2%}'.format)
data['個人情報%'] = (data['個人情報'] / data['卒業年度']).map('{:.2%}'.format)
```
